```python
from bs4 import BeautifulSoup

# file path same as before i used
file_path = "unlocked-lg-6in-128gb-factoryunlocked-used.html"
with open(file_path, "r", encoding="utf-8") as file:
    html_content = file.read()

# parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# get all div element with the class is s-item__info
info = soup.find_all("div", class_="s-item__info")

# loop to print out
for i in info:
    # get title in div with the class is s-item__title
    title_element = i.find("div", class_="s-item__title")
    title = title_element.text.strip() if title_element else "N/A"

    # get seller in span with the class is s-item__seller-info-text
    seller_element = i.find("span", class_="s-item__seller-info-text")
    seller = seller_element.text.strip() if seller_element else "N/A"

    # get seller_rating in span with the class is clipped
    seller_rating_element = i.find("span", class_="clipped")
    seller_rating = seller_rating_element.get("title") if seller_rating_element else "N/A"

    # get price in span with the class is s-item__price
    price_element = i.find("span", class_="s-item__price")
    price = price_element.text.strip() if price_element else "N/A"

    # get shipping_info in span with the class is s-item__shipping
    shipping_element = i.find("span", class_="s-item__shipping")
    shipping_info = shipping_element.text.strip() if shipping_element else "N/A"

    # get return_info in span with the class is s-item__free-returns s-item__freeReturnsNoFee
    return_element = i.find("span", class_="s-item__free-returns s-item__freeReturnsNoFee")
    return_info = return_element.text.strip() if return_element else "N/A"

    # get bids in span with the class is s-item__bids s-item__bidCount
    bids_element = i.find("span", class_="s-item__bids s-item__bidCount")
    bids = bids_element.text.strip() if bids_element else "N/A"

    # printout
    print("Title:", title)
    print("Seller:", seller)
    print("Seller Rating:", seller_rating)
    print("Price:", price)
    print("Shipping Information:", shipping_info)
    print("Return Information:", return_info)
    print("Number of Bids:", bids)
    print("\n")

```
